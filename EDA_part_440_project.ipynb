```python
import pandas as pd

file_path = "/content/drive/MyDrive/CSE440 Project DataSet/Question Answer Classification Dataset 1[Training].csv"
df = pd.read_csv(file_path)

test_file_path = "/content/drive/MyDrive/CSE440 Project DataSet/[Updated] Question Answer Classification Dataset[Test].csv"
test_df = pd.read_csv(test_file_path)

print("Training DataFrame:")
print("Shape:", df.shape)
print("Column Names:", df.columns)
print("Data Types:\n", df.dtypes)
print("\nTesting DataFrame:")
print("Shape:", test_df.shape)
print("Column Names:", test_df.columns)
print("Data Types:\n", test_df.dtypes)
```

## Analyze 'class' distribution

### Subtask:
Investigate the distribution of the target variable 'Class' in both the training and testing datasets to check for class imbalance.


**Reasoning**:
Calculate and print the value counts for the 'Class' column in both the training and testing dataframes to investigate class distribution.



```python
train_class_distribution = df['Class'].value_counts()
test_class_distribution = test_df['Class'].value_counts()

print("Training Class Distribution:")
print(train_class_distribution)
print("\nTesting Class Distribution:")
print(test_class_distribution)
```

## Analyze 'qa text' content

### Subtask:
Explore characteristics of the text data in the 'QA Text' column, such as text length, word counts, and potentially the most frequent words (after basic tokenization or cleaning).


**Reasoning**:
Calculate and display the text length and word count for the 'QA Text' column in both dataframes and then print descriptive statistics for these new columns.



```python
df['text_length'] = df['QA Text'].apply(len)
df['word_count'] = df['QA Text'].apply(lambda x: len(x.split()))

test_df['text_length'] = test_df['QA Text'].apply(len)
test_df['word_count'] = test_df['QA Text'].apply(lambda x: len(x.split()))

print("Training Data - Text Length and Word Count Statistics:")
print(df[['text_length', 'word_count']].describe())

print("\nTesting Data - Text Length and Word Count Statistics:")
print(test_df[['text_length', 'word_count']].describe())
```

## Check for missing values

### Subtask:
Confirm there are no missing values in either dataset (as indicated by `df.info()` and `test_df.info()`, but it's good practice to explicitly check).


**Reasoning**:
Check for missing values in the training and testing dataframes.



```python
print("Missing values in training data:")
print(df.isnull().sum())

print("\nMissing values in testing data:")
print(test_df.isnull().sum())
```

## Visualize data

### Subtask:
Create visualizations to represent the findings from the analysis, such as bar plots for class distribution and histograms for text length.


**Reasoning**:
Create visualizations to represent the class distribution, text length, and word count for both training and testing datasets using bar plots and histograms.



```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.barplot(x=train_class_distribution.index, y=train_class_distribution.values)
plt.title("Training Class Distribution")
plt.xlabel("Class")
plt.ylabel("Count")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 6))
sns.barplot(x=test_class_distribution.index, y=test_class_distribution.values)
plt.title("Testing Class Distribution")
plt.xlabel("Class")
plt.ylabel("Count")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 6))
sns.histplot(df['text_length'], bins=50, kde=True)
plt.title("Distribution of Text Length in Training Data")
plt.xlabel("Text Length")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 6))
sns.histplot(test_df['text_length'], bins=50, kde=True)
plt.title("Distribution of Text Length in Testing Data")
plt.xlabel("Text Length")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 6))
sns.histplot(df['word_count'], bins=50, kde=True)
plt.title("Distribution of Word Count in Training Data")
plt.xlabel("Word Count")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 6))
sns.histplot(test_df['word_count'], bins=50, kde=True)
plt.title("Distribution of Word Count in Testing Data")
plt.xlabel("Word Count")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```
